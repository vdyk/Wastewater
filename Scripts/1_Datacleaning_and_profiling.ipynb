# 04. Data Cleaning

```python
# dropping the columns that are not needed
df_merged.drop(['country_y', 'city_y', 'location', 'institution', 'website'], axis=1, inplace=True)
```

```python
#renaming columns
df_merged.rename(columns={'country_x': 'Country', 'city_x': 'City'}, inplace=True)
```

```python
#changing datatype for latitude
df_merged['latitude'] = df_merged['latitude'].astype(float)
```

```python
df_merged.info()
```

# 05. Consistency checks

```python
#preparing descriptive statistics
df_merged.describe()
```

```python
# frequency tables for categorical variables
pd.set_option("display.max_rows", None)
print(df_merged.Country.value_counts())
print(df_merged.City.value_counts())
print(df_merged.SiteID.value_counts())
print(df_merged.metabolite.value_counts())
```

```python
# looking for mixed datatypes
for col in df_merged.columns.tolist():
  mixed = (df_merged[[col]].applymap(type) != df_merged[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_merged[mixed]) > 0:
    print (col)
```

```python
# looking for missing values
df_merged.isnull().sum()
```

```python
#looking for duplicate values
df_dups= df_merged[df_merged.duplicated()]
print(df_dups)
```
